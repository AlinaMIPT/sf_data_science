# Преобразование данных

1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
def get_education (education):
    education_list = education.split(' ')
    education_short_list = education_list[:2]
    if education_short_list[0] == "Высшее":
        education_type = "высшее"
    elif education_short_list[0] == "Неоконченное":
        education_type = "неоконченное высшее"
    elif education_short_list[0] == "Среднее":
        if education_short_list[1] == "специальное":
            education_type = "среднее специальное"
        else:
            education_type = "среднее"
    return education_type

hh_data['Образование'] = hh_data['Образование и ВУЗ'].apply(get_education)
hh_data[hh_data['Образование'] == 'среднее'].shape[0]
hh_data.drop("Образование и ВУЗ", axis = 1, inplace = True)
```

2. Теперь нас интересует столбец **"Пол/возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
def get_gender (sex):
    sex_list = sex.split(' , ')
    sex_short_list = sex_list[:1]
    if sex_short_list[0] == "Мужчина":
        sex_type = "М"
    else:
        sex_type = "Ж"
    return sex_type
def get_age (age):
    age_list = age.split(' , ')
    age_short_list = age_list[1]
    age_short_list_new = age_short_list.split(' ')
    age_number = age_short_list_new[1]
    return age_number
hh_data['Пол'] = hh_data['Пол, возраст'].apply(get_gender)
hh_data['Возраст'] = hh_data['Пол, возраст'].apply(get_age)
round(hh_data['Пол'].value_counts(normalize=True)*100,2)
hh_data['Возраст'] = hh_data['Возраст'].astype('int')
round(hh_data['Возраст'].mean(),1)
hh_data.drop("Пол, возраст", axis = 1, inplace = True)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-третьих, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_experience(experience):
    months_keywords = ['месяц','месяца','месяцев']
    years_keywords = ['года','лет','год']
    month = 0
    year = 0
    if experience is not np.nan:
        exp_split = experience.split(' ')
        exp_split_new = exp_split[:7]
        for i in range(len(exp_split_new)):
            if exp_split_new[i] in months_keywords:
                month = exp_split_new[i-1]
            if exp_split_new[i] in years_keywords:
                year = exp_split_new[i-1]
        return int(year)*12+int(month)
    else:
        return np.nan  
hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(get_experience)
hh_data['Опыт работы (месяц)'].median()
hh_data.drop("Опыт работы", axis = 1, inplace = True)
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#определение города
def get_city (city):
    million_cities = ['Сочи','Москва','Санкт-Петербург','Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск',
                      'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    full_list = city.split(' , ')
    city = full_list[0]
    if city not in million_cities:
        city = 'другие'
    return city
hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(get_city)

#определение готовности к перезду
def get_move_status (moving):
    ready_keywords = ['готов','готова','хочу']
    move_split = moving.split(' , ')
    if len(move_split) == 4:
        move_status = move_split[2]
    else:
        move_status = move_split[1]
    move_status_split = move_status.split(' ')
    for i in range(len(move_status_split)):
        if move_status_split[i] in ready_keywords:
            if move_status_split[i-1] == 'не':
                return False
            else:
                return True
        else:
            return False
hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(get_move_status)

#определение готовности к командировке
def get_business_trip_status (trip):
    ready_keywords = ['готов','готова']
    trip_split = trip.split(' , ')
    if len(trip_split) == 4:
        trip_status = trip_split[3]
    elif len(trip_split) == 3:
        trip_status = trip_split[2]
    elif len(trip_split) == 5:
        trip_status = trip_split[4]
    else:
        trip_status = '' 
    if trip_status is '':
        return False
    else:
        trip_status_split = trip_status.split(' ')
        for i in range(len(trip_status_split)):
            if trip_status_split[i] in ready_keywords:
                if trip_status_split[i-1] == 'не':
                    return False
                else:
                    return True
            else:
                return False
hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(get_business_trip_status)

#выполнение заданий
round(hh_data[hh_data['Город'] == "Санкт-Петербург"].shape[0]/hh_data.shape[0]*100,0)
round(hh_data[(hh_data['Готовность к командировкам'] == True) & (hh_data['Готовность к переезду'] == True)].shape[0]/hh_data.shape[0]*100,0)

#удаление столбца "Город, переезд, командировки" 
hh_data.drop("Город, переезд, командировки", axis = 1, inplace = True)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
def get_split_occupation_schedule (occupation):
    occupation_split = occupation.split(', ')
    return occupation_split

hh_data['Занятость'] = hh_data['Занятость'].apply(get_split_occupation_schedule)
hh_data['полная занятость'] = hh_data['Занятость'].apply(lambda x:'полная занятость' in x)
hh_data['частичная занятость'] = hh_data['Занятость'].apply(lambda x:'частичная занятость' in x)
hh_data['проектная работа'] = hh_data['Занятость'].apply(lambda x:'проектная работа' in x)
hh_data['стажировка'] = hh_data['Занятость'].apply(lambda x:'стажировка' in x)
hh_data['волонтерство'] = hh_data['Занятость'].apply(lambda x:'волонтерство' in x)

hh_data['График'] = hh_data['График'].apply(get_split_occupation_schedule)
hh_data['полный день'] = hh_data['График'].apply(lambda x:'полный день' in x)
hh_data['сменный график'] = hh_data['График'].apply(lambda x:'сменный график' in x)
hh_data['гибкий график'] = hh_data['График'].apply(lambda x:'гибкий график' in x)
hh_data['удаленная работа'] = hh_data['График'].apply(lambda x:'удаленная работа' in x)
hh_data['вахтовый метод'] = hh_data['График'].apply(lambda x:'вахтовый метод' in x)

#выполнение задания
hh_data[(hh_data['проектная работа'] == True) & (hh_data['волонтерство'] == True)].shape[0]
hh_data[(hh_data['вахтовый метод'] == True) & (hh_data['гибкий график'] == True)].shape[0]

#удаление столбцов
hh_data.drop("Занятость", axis = 1, inplace = True)
hh_data.drop("График", axis = 1, inplace = True)
```

```python
#получение необходимого dataframe для конвертирования валют
exchange_rates = pd.read_csv('ExchangeRates_Project_1.ipynb')
exchange_rates["date"] = pd.to_datetime(exchange_rates["date"])
exchange_rates_df = exchange_rates[['currency','date','proportion','close']]
exchange_rates_df['date'] = exchange_rates_df['date'].dt.date

#функции для получения валюты и зарплаты (только число)
def get_currency(salary):
    salary_split = salary.split(' ')
    currency = salary_split[1]
    return currency
def get_salary_number (salary):
    salary_split = salary.split(' ')
    salary_number = salary_split[0]
    return salary_number

#приведение столбцов к необходимому формату
hh_data['Обновление резюме'] = pd.to_datetime(hh_data['Обновление резюме'])
hh_data['date'] = hh_data['Обновление резюме'].dt.date
hh_data['currency'] = hh_data['ЗП'].apply(get_currency)
hh_data.loc[hh_data['currency'] == 'грн.','currency'] = 'UAH'
hh_data.loc[hh_data['currency'] == 'бел.руб.','currency'] = 'BYN'
hh_data.loc[hh_data['currency'] == 'сум','currency'] = 'UZS'

#объединение таблиц
hh_data_merged = hh_data.merge(
    exchange_rates_df,
    on=['currency','date'],
    how='left'
)

hh_data_merged.loc[hh_data_merged['currency'] == 'руб.','proportion'] = 1
hh_data_merged.loc[hh_data_merged['currency'] == 'руб.','close'] = 1

hh_data_merged['ЗП (руб)'] = hh_data_merged['ЗП'].apply(get_salary_number)
hh_data_merged['ЗП (руб)'] = hh_data_merged['ЗП (руб)'].astype('int64')
hh_data_merged['ЗП (руб)'] = hh_data_merged['ЗП (руб)'] * hh_data_merged.close / hh_data_merged.proportion

#выполнение задания
hh_data_merged['ЗП (руб)'].median()

#удаление столбца ЗП
hh_data_merged.drop("ЗП", axis = 1, inplace = True)
```

Найдите выбросы с помощью **метода z-отклонения** и удалите их из данных, используйте логарифмический масштаб, сделайте **послабление на 1 сигму (возьмите 4 сигмы)** в сторону, противоположную асимметрии. 

```python
#гистограмма в логарифмическом масштабе
log_age= np.log(hh_data_merged['Возраст'])
histplot = sns.histplot(log_age , bins=30)
histplot.set_title('Распределение возраста в логарифмическом масштабе');

#логарифмическое рапределение асимметрично слегка влево

def outliers_z_score_mod(data, feature, log_scale=False,left=3,right=3):
    if log_scale:
        x = np.log(data[feature])
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_z_score_mod(hh_data_merged, 'Возраст', log_scale=True,left=3,right=4)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
```
